```python
import numpy as np
#%matplotlib notebook # untuk menampilkan gambar dalam mode windows figure
#untuk menampilkan gambar secara langsung di web coding
%matplotlib inline 

x = np.linspace(0, 50, 100)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.show()
```

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)

fig, ax = plt.subplots()
plt.style.use('seaborn')
plt.plot(x,y,x,y2,x,y3)
plt.show()
```

```python
#Multiple plots with single axes
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
sin2= np.sin(x)*2
cos = np.cos(x)
cos3 = np.cos(x)*2

fig, ax = plt.subplots (nrows=2, ncols=2, figsize=(12,8), sharey=True, sharex=True)
ax[0][0].plot(x, sin, label='sin (x - 90)', color='b')
ax[0][1].plot(x, sin2, label='2sin x', color='r')
ax[1][0].plot(x, cos, label='cos x', color='y')
ax[1][1].plot(x, cos3, label='3cos x', color='g')

fig.legend()
plt.show()
```

```python
#Nested Axis
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)

fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x, y, color='b', label='cos x')

ins1 = ax.inset_axes([-0.2, -0.2, 0.3, 0.3])
ins1.plot(x, y2, color='r', label = 'cosx^2')
ins1.plot(x, y, color='b')

ins2 = ax.inset_axes([0.8, -0.2, 0.3, 0.3])
ins2.plot(x, y3, color='g', label = 'cosx^3')
ins2.plot(x, y, color='b')

ins3 = ax.inset_axes([0.8, 0.8, 0.3, 0.3])
ins3.plot(x, y4, color='y', label='cosx^4')
ins3.plot(x, y, color='b')

ins4 = ax.inset_axes([-0.2, 0.8, 0.3, 0.3])
ins4.plot(x, y5, color='orange', label = 'cosx^5')
ins4.plot(x, y, color = 'b')

ax.set_title('Grafik cosx')
ins1.set_title('Grafik cosx^2')
ins2.set_title('Grafik cosx^3')
ins3.set_title('Grafik cosx^4')
ins4.set_title('Grafik cosx^5')

plt.show()
```

### Histogram, Binnings and Density

```python
#Histogram, Binnings and Density
x1 = np.random.normal(-5, 7, 1000)
x2 = np.random.normal(6, 4, 1000)
x3 = np.random.normal(0, 5, 1000)

hists = dict(histtype='stepfilled', alpha=0.5, bins=30)

fig, ax = plt.subplots(figsize = (15,9))
plt.hist(x1, **hists)
plt.hist(x2, **hists)
plt.hist(x3, **hists)

plt.show()
```

```python
data = pd.Series([56,25, 5, 131, 232, 443, 689, 989, 229, 75, 12, 5, 90, 134, 178, 133, 201, 256,
                 234, 386, 319, 358, 400, 489, 422, 470, 587, 523, 511, 505, 617, 604, 688, 643, 782, 724, 763,700,
                 801, 839, 850, 875, 897, 910, 924, 957, 983, 949, 917])

counts, bin_edges = np.histogram(data, bins=15)
print(counts)
print(bin_edges)

mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 1000). T

# A two dimmensional histogram with plt.hist2d
fig, ax = plt.subplots(figsize=(10, 6))
plt.hist2d(x, y, bins=35, cmap = 'Blues')
duadimensi = plt.colorbar()
duadimensi.set_label('count in bins')
```

```python
# plt.hexbin: Hexagonal binnings
fig, ax = plt.subplots(figsize=(10, 6))
plt.hexbin(x, y, gridsize=35, cmap= 'Blues')
cb = plt.colorbar(label= 'count in bin')
```

Another common method of evaluating densities in multiple dimensions is kernel
density estimation (KDE). KDE can be thought of as a way to “smear out” the points in space and add up the result to obtain a smooth function.

```python
#Kernel density estimation
from scipy.stats import gaussian_kde

## fit an array of size [Ndim, Nsamples]
datakernel = np.vstack([x, y])
kde = gaussian_kde(datakernel)

## evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z= kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

## plot the result as an image
fig, ax = plt.subplots(figsize=(10, 6))
plt.imshow(Z.reshape(Xgrid.shape), 
          origin= 'lower', aspect='auto', extent = [-3.5, 3.5, -6, 6],
          cmap='Blues')
kde = plt.colorbar()
kde.set_label('density')
```

<b>Let's start off by analyzing the effect of Iceland's Financial Crisis</b>:

The 2008 - 2011 Icelandic Financial Crisis was a major economic and political event in Iceland. Relative to the size of its economy, Iceland's systemic banking collapse was the largest experienced by any country in economic history. The crisis led to a severe economic depression in 2008 - 2011 and significant political unrest.

Question: Let's compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.
